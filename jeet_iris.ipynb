```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#we will load csv file into a pandas dataFrame.
iris = pd.read_csv("iris.csv")
```

```python
#Different Plots
```

```python
# 2-D Scatter plot
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue="Species", size=4) \
   .map(plt.scatter, "SepalLength", "SepalWidth") \
   .add_legend();
plt.show();
```

```python
#conclusion from 2d scatter plot
#The plot between SepalLength and SepalWidth shows clearly shows that setosa flowers can be easily separated, 
#but it is very hard to seperate versicolor from virginicia.
```

```python
#Next thing is lets apply pair plot and explore the dataset
plt.close();
sns.set_style("whitegrid");
sns.pairplot(iris, hue="Species", size=3);
plt.show()
```

```python
#Concluding remarks
#So with the above pair plots we observe that PetalLength and PetalWidth are the most important features to distinguish
#these flower types && Setosa can be easily identified (linearly seperable), Versicolor and Virginicia have some overlap.
```

```python
import numpy as np
jitendra = pd.read_csv("setosa.csv")
```

```python
#CDF of petal_length
counts, bin_edges = np.histogram(jitendra['PetalLength'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)



plt.show();
```

```python
#In this figure CDF tells us about the percentage of flowers to corresponding length, like in the above graph suppose we take 
#0.80, marking this will lead to a point somewhere around 1.6. it means 80% of the setosa flower have petal length of <1.6 and so on.
```
